# Lunar Lander

```python
import ray
from ray import tune
from ray.rllib import agents
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Tuning Hyperparameters

```python
results = tune.run(
    'DQN', 
    stop={
        'timesteps_total': 50000
    },
    config={
    "env": 'LunarLander-v2',
    "num_workers": 3,
    "gamma" : tune.grid_search([0.999, 0.8]),
    "lr": tune.grid_search([1e-2, 1e-3, 1e-4]),
    }
)
```

### Finding out best parameters combination

```python
print("Best config is", results.get_best_config(metric="episode_reward_min"))
```

```python
df = results.dataframe()
print(df.columns)
df.head()
```

### Plotting Metrics

```python
plt.figure(num=None, figsize=(8, 5), dpi=80, facecolor='w', edgecolor='k')
legends = []
for _ , path_dir in df['logdir'].iteritems():
    data = pd.read_csv(path_dir + '/progress.csv')
    gamma = data['experiment_tag'][0].split('=')[1].split(',')[0]
    lr = data['experiment_tag'][0].split('=')[-1]
    legends.append('Gamma:'+ str(gamma) + ', LR:' + str(lr))
    
    plt.plot(data['timesteps_total'], data['episode_reward_mean'])
    
plt.xlabel('Time Steps')
plt.ylabel('Averaged Rewards')
plt.title('Training Rewards over Time')
plt.legend(labels=legends)
plt.savefig("rewards.png", format='png', dpi=1200)
plt.show()
```
