# MESA - Agent Based Modelling (Forest Fire Simulation)

```python
from mesa import Model
from mesa.datacollection import DataCollector
from mesa.time import RandomActivation
from mesa import Agent
from mesa.space import MultiGrid
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
def grid_stats(model, it):
    # Reference: https://dmnfarrell.github.io/bioinformatics/abm-mesa-python
    df=pd.DataFrame(np.zeros((model.grid.width, model.grid.height)))
    for (states, x, y) in model.grid.coord_iter():
        c=None
        for i in states:
            c = i.condition
        df.iloc[x,y] = c
    df = pd.DataFrame(df.stack(), columns=['value']).reset_index()
    df['iter'] = [it]*len(df)
    return df

def trees_stats(model, tree_condition):
    return sum([1 for tree in model.schedule.agents if tree.condition == tree_condition])
```

```python
class Tree(Agent):
    def __init__(self, pos, model):
        super().__init__(pos, model)
        self.pos = pos
        self.condition = "Healthy"

    def step(self):
        if self.condition == "Burning":
            for close_tree in self.model.grid.neighbor_iter(self.pos):
                if close_tree.condition == "Healthy":
                    close_tree.condition = "Burning"
            self.condition = "Dead"

    def get_pos(self):
        return self.pos
```

```python
class ForestModel(Model):
    def __init__(self, height, width, prob):
        self.height, self.width = height, width
        self.prob = prob
        self.schedule = RandomActivation(self)
        self.grid = MultiGrid(width, height, True)
        self.datacollector = DataCollector({
                "Healthy": lambda m: trees_stats(m, "Healthy"),
                "Burning": lambda m: trees_stats(m, "Burning"),
                "Dead": lambda m: trees_stats(m, "Dead")})

        for (states, x, y) in self.grid.coord_iter():
            if self.random.random() < self.prob:
                new_tree = Tree((x, y), self)
                if x == 0:
                    new_tree.condition = "Burning"
                self.grid._place_agent((x, y), new_tree)
                self.schedule.add(new_tree)

        self.running = True
        self.datacollector.collect(self)

    def step(self):
        self.schedule.step()
        self.datacollector.collect(self)
        if trees_stats(self, "Burning") == 0:
            self.running = False
```

```python
steps=20
model = ForestModel(100, 100, 0.5)
res = pd.DataFrame([])
for i in range(steps):
    model.step()    
    data = grid_stats(model, i)
    res = res.append(data)
    
res2 = model.datacollector.get_model_vars_dataframe()
```

```python
res2['Iteration'] = list(res2.index)
res2 = res2[:len(res2)-1]
ind_arr = [0]
for i in res2['Iteration'].unique()[1:]:
    ex = res2[res2['Iteration']==i-1].values
    indn = 0
    for j in range(len(ex)):
        ex[j][-1] = i
        res2.loc[len(res2)] = ex[j]
        ind_arr.append(indn)
        indn += 1
    ind_arr.append(indn)
res2 = res2.sort_values(by=['Iteration', 'Healthy'], ascending=[True, False])
res2.reset_index(drop=True, inplace=True)
res2['index'] = ind_arr

res.columns = ['X', 'Y', 'Status', 'Iteration']
```

```python
px.line(res2,
        title="Forest Fire Simulation",
        width=700, height=400,
        x='index', y=res2.columns[0:-1], animation_frame='Iteration',  range_x=[0,20])
```
