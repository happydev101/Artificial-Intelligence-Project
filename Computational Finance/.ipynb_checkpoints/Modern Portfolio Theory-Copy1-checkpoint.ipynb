```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.seasonal import seasonal_decompose
from yahoofinancials import YahooFinancials
import scipy.optimize as sco
```

## Modern Portfolio Theory

```python
df = YahooFinancials(['JPM', 'GS', 'HSBC']).get_historical_price_data(start_date='2017-01-01', 
                                                                       end_date='2019-12-31', 
                                                                       time_interval='daily')
df = pd.DataFrame({'Date' :pd.DataFrame(df['JPM']['prices'])['formatted_date'],
                   'JPM' :pd.DataFrame(df['JPM']['prices'])['adjclose'],
                   'GS' :pd.DataFrame(df['GS']['prices'])['adjclose'],
                   'HSBC' :pd.DataFrame(df['HSBC']['prices'])['adjclose']})
df = df.set_index('Date')
df.head()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
df['JPM'].plot();
df['GS'].plot();
df['HSBC'].plot();
plt.legend(['JPM', 'GS', 'HSBC'])
plt.xlabel("Date", fontsize=18);
plt.ylabel("Price", fontsize=18);
plt.title('Portfolio Stock Prices over time', fontsize=18);
plt.savefig("4.png", format='png', dpi=1200, bbox_inches='tight')
```

```python
def portfolio_sim(df, daily_mean_returns, daily_cov_matrix, rf_rate, itern, working_days=252):
    number_of_stocks = df.shape[1]
    sim_res = np.zeros((3, itern))
    list_weights = np.empty((0, number_of_stocks))
    for i in range(itern):
        random_weights = np.random.random(number_of_stocks)
        normalized_weights = random_weights/np.sum(random_weights)
        list_weights = np.append(list_weights, [normalized_weights], axis=0)
        port_annual_return = np.sum(daily_mean_returns*normalized_weights) * working_days 
        port_annual_std = np.sqrt(np.dot(normalized_weights.T, np.dot(daily_cov_matrix, 
                                                           normalized_weights))) * np.sqrt(working_days)
        sim_res[0,i], sim_res[1,i] = port_annual_std, port_annual_return
        sim_res[2,i] = (port_annual_return - rf_rate) / port_annual_std
    
    min_risk, max_sharpe = np.argmin(sim_res[0]), np.argmax(sim_res[2])
    
    def min_std(normalized_weights, daily_cov_matrix, working_days):
        return np.sqrt(np.dot(normalized_weights.T, np.dot(daily_cov_matrix, 
                                                           normalized_weights))) * np.sqrt(working_days)
    
    r_range = np.linspace(sim_res[1, min_risk], sim_res[1, max_sharpe], 70)
    w_dist = [1/number_of_stocks for i in range(number_of_stocks)]
    effective_p = []
    for r in r_range:
        const = ({'type': 'eq', 'fun': lambda x: np.sum(daily_mean_returns*x) * working_days - r},
                   {'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
        limits = tuple((0,1) for asset in range(number_of_stocks))
        res = sco.minimize(min_std,  w_dist, args=(daily_cov_matrix, working_days), 
                              method='SLSQP', bounds=limits, constraints=const)
        effective_p.append(res['fun']*100)
    
    plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.scatter(sim_res[0,:]*100, sim_res[1,:]*100, c=sim_res[2,:]*100, cmap='viridis', marker='.')
    plt.colorbar(orientation='vertical')
    plt.scatter(sim_res[0, min_risk]*100, sim_res[1, min_risk]*100, 
                marker='x', color='r', s=300, label='Minimum Risk')
    plt.scatter(sim_res[0, max_sharpe]*100, sim_res[1, max_sharpe]*100, 
                marker='x', color='b', s=300, label='Maximum Sharpe Ratio')
    plt.plot(effective_p, r_range*100, color='black', label='Effective Frontier')
    plt.legend(labelspacing=0.7)
    
    plt.title('Monte Carlo Simulated Effective Frontier', fontsize=18)
    plt.xlabel('Portfolio Standard Deviation', fontsize=18)
    plt.ylabel('Portfolio Expected Return', fontsize=18)
    
    plt.savefig("5.png", format='png', dpi=1200, bbox_inches='tight')
    plt.show()
    df2 = pd.DataFrame([list_weights[min_risk], list_weights[max_sharpe]], 
                       columns=df.columns, index=["Minimum Risk Portfolio", 
                                                  "Maximum Sharpe Portfolio"])
    return df2
```

```python
daily_returns = df.pct_change()
df2 = portfolio_sim(df, daily_returns.mean(), daily_returns.cov(), 0.028, 5000)
pd.set_option("display.precision", 3)
df2.head()
```
